## Algoritmo de Ordenamiento por Inserción

Este algoritmo de ordenamiento toma un par de elementos y los compara para organizarlos de manera ascendente (en este caso). Entre más aumenta el ciclo externo más interacciones se deben realizar en el ciclo interno para verificar el orden de todos los elementos.

```python
def ordenamientoInsercion(array): 

  for i in range(1, len(array)): 

    for j in range(i, 0, -1):
      # entre mas grande sea i mas interacciones tendra este loop
      elemento = array[j];
      elemento_anterior = array[j-1];
      if elemento < elemento_anterior:
        # si se cumple la condicion, simplemente intercambiamos posiciones
        array[j], array[j-1] = elemento_anterior, elemento;
        
  return 
```

```python
lista = [22,14,10,5,6,2,1,2,21,13]

ordenamientoInsercion(lista)
print(f"Lista Organizada:{lista}")
```
