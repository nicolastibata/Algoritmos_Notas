## Algoritmo de Ordenamiento por Selección

Este algoritmo de ordenamiento toma el primer elemento del arreglo y lo compara con los demas elementos. En caso de que encuentre un elemento menor al primer elemento, entonces se intercambian. De tal forma se realiza con cada uno de los elementos a travez del length() y asi se organizan todos los elementos.

```python
def ordenamientoSeleccion(array): ## de manera ascendente

  for i in range(0, len(array)): 
    # Un loop que empieza en 0 y termina un numero antes que el length de la lista
    elemento_menor = array[i];
    posicion_menor = i;
    # Asumimos que el menor elemento es el que esta en la primera posicion de la lista

    for j in range(i+1, len(array)):
      # El loop interno empieza una posicion siguiente al "menor_elemento"
      elemento_siguiente = array[j];
      if elemento_menor > elemento_siguiente:
        # Si el elemento menor (el que asumimos) es mayor al elemento siguiente, entonces se cambian las variables
        elemento_menor = elemento_siguiente;
        posicion_menor = j;
        # Y asi sucesivamente hasta encontrar el menor elemento de tooda la lista
    array[i], array[posicion_menor] = elemento_menor, array[i]
    # Al terminar el loop interno entonces cambiamos variables, con la seguridad de que es el elemento mas pequeño
    
  return 
```

```python
lista = [22,14,10,5,6,2,1,2,21,13]

ordenamientoSeleccion(lista)
print(f"Lista Organizada:{lista}")
```
